```python
import pandas as pd

# Load CSVs
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
train.info()
train.isnull().sum()
```

```python
for col in ['Gender', 'Married', 'Dependents', 'Self_Employed']:
    train[col].fillna(train[col].mode()[0], inplace=True)
    test[col].fillna(test[col].mode()[0], inplace=True)
```

```python
train['LoanAmount'].fillna(train['LoanAmount'].median(), inplace=True)
test['LoanAmount'].fillna(test['LoanAmount'].median(), inplace=True)
train['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].mode()[0], inplace=True)
test['Loan_Amount_Term'].fillna(test['Loan_Amount_Term'].mode()[0], inplace=True)

```

```python

train['Credit_History'].fillna(train['Credit_History'].mode()[0], inplace=True)
test['Credit_History'].fillna(test['Credit_History'].mode()[0], inplace=True)
```

```python
train['Dependents'].fillna('0', inplace=True)
test['Dependents'].fillna('0', inplace=True)
```

```python
train['Dependents'] = train['Dependents'].replace('3+', '3+').astype(str)
test['Dependents'] = test['Dependents'].replace('3+', '3+').astype(str)
```

```python
cat_cols = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area']

for col in cat_cols:
    train[col] = train[col].astype(str)
    test[col] = test[col].astype(str)
```

```python
train['Gender'] = train['Gender'].map({'0': 'Female', '1': 'Male'})
test['Gender'] = test['Gender'].map({'0': 'Female', '1': 'Male'})
```

```python
train['Married'] = train['Married'].map({'0': 'No', '1': 'Yes'})
test['Married'] = test['Married'].map({'0': 'No', '1': 'Yes'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python

sns.countplot(x='Loan_Status', data=train)
plt.title('Loan Status Distribution')
plt.show()
```

```python
cat_features = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Dependents']

for col in cat_features:
    sns.countplot(x=col, hue='Loan_Status', data=train)
    plt.title(f'{col} vs Loan_Status')
    plt.show()
```

```python
num_features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

train[num_features].hist(figsize=(10,8), bins=20)
plt.show()

for col in num_features:
    sns.boxplot(x='Loan_Status', y=col, data=train)
    plt.title(f'{col} vs Loan_Status')
    plt.show()
```

```python

train['Loan_Status'] = train['Loan_Status'].map({'Y': 1, 'N': 0})
```
